```python
import numpy as np
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures

train_data = []
test_data = []
for i in range(432):
    train_data.append(l[i])
for i in range(432,480):
    test_data.append(l[i])

Y = []
for i in range(480):
    Y.append(l[i][-1])

sum_Y = sum(Y)
mean_Y = sum_Y/480

S_total = 0 
for i in range(480):
    S_total += (Y[i]-mean_Y)**2
    
train_data = np.array(train_data)
test_data = np.array(test_data)

X_train = train_data[:,0:7]
Y_train = train_data[:,-1]

X_test = test_data[:,0:7]
Y_actual = test_data[:,-1]

poly = PolynomialFeatures(degree=2)
X_train = poly.fit_transform(X_train)
X_test = poly.fit_transform(X_test)

model = linear_model.LinearRegression()
model.fit(X_train,Y_train)
Y_test = model.predict(X_test)
result = '\n'.join(list(map(str,Y_test)))

S_res = 0
for i in range(len(Y_actual)):
    S_res += (Y_actual[i]-Y_test[i])**2

print("R^2 Value: ",(1-(S_res/S_total))*100)

#print(result)
```
